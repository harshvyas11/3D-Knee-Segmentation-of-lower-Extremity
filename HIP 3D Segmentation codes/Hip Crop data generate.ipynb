```python
def read_csv_to_dict_LM_INFO(csv_file):
    LM_INFO = {}
    with open(csv_file, mode='r') as infile:
        reader = csv.reader(infile)       
        for rows in reader:
            LM_INFO[rows[0]] = np.asarray([float(rows[1]), float(rows[2]), float(rows[3])])
            #print(rows[0], ':', float(rows[1]), float(rows[2]), float(rows[3]))
            
    return LM_INFO
```

```python
def zoom_1mm(img,original_space,new_space=[1.0,1.0,1.0]):
    zoom_factor = [original_space[i] / new_space[i] for i in range(3)]
    zoomed_image = zoom(img, zoom_factor, order=0)
    
    return zoomed_image
```

```python
def hip_centre(path, original_space):
    HC = read_csv_to_dict_LM_INFO(path)['FHP']
    HC_1mm = [round(original_space[i] * HC[i]) for i in range(3)]
    
    return HC_1mm
```

```python
def CROP_3D_CT_IMAGE_AT_GIVEN_POINT_NEW_256(img, centered_at, crop_size):
    crop_size = np.asanyarray(crop_size)
    img = np.pad(img, ((128,128),(128,128),(128,128)),'edge')   

    sp_idx = np.asarray(128+centered_at-crop_size//2,dtype=int)
    ep_idx = np.asarray(128+centered_at+crop_size//2,dtype=int)
       
    return img[sp_idx[0]:ep_idx[0], sp_idx[1]:ep_idx[1], sp_idx[2]:ep_idx[2]]
```

```python
crop_hip_train = 'F:\\Birva_Harsh\\Hip_Crop_Data\\train\\Hip'
crop_mask_train = 'F:\\Birva_Harsh\\Hip_Crop_Data\\train\\Mask'
```

```python
for i in range(len(df_train)):
    img, kinfo = nrrd.read(df_train.loc[i,'hip'])
    mask, minfo = nrrd.read(df_train.loc[i,'mask'])
    
    spc_dir = np.diag(kinfo['space directions'])
    
    img_1mm = zoom_1mm(img, spc_dir)
    mask_1mm = zoom_1mm(mask, spc_dir)
    
    HC = hip_centre(csv_paths[i], spc_dir)
    
    kinfo['sizes'] = np.array([256.0,256.0,256.0])
    minfo['sizes'] = np.array([256.0,256.0,256.0])
    kinfo['space directions'] = np.eye(3)
    minfo['space directions'] = np.eye(3)
    
    for j in range(5):
        print(f"{os.path.basename(os.path.dirname(df_train.loc[i,'hip']))}_{j+1}")
        shift = np.random.randint(-30,30,size=3)
        shift_HC = HC - shift
        
        crop_hip = CROP_3D_CT_IMAGE_AT_GIVEN_POINT_NEW_256(img_1mm, shift_HC, [256,256,256])
        crop_mask = CROP_3D_CT_IMAGE_AT_GIVEN_POINT_NEW_256(mask_1mm, shift_HC, [256,256,256])
        
        output_img_path = os.path.join(crop_hip_train, f"{os.path.basename(os.path.dirname(df_train.loc[i,'hip']))}_{j+1}.nrrd")
        output_mask_path = os.path.join(crop_mask_train, f"{os.path.basename(os.path.dirname(df_train.loc[i,'hip']))}_{j+1}.nrrd")
        
        nrrd.write(output_img_path, crop_hip, kinfo)
        nrrd.write(output_mask_path, crop_mask, minfo)
```

```python
def plot_3d(img,mask):
    plt.figure(figsize=(15,8))
    plt.subplot(2, 3, 1)
    plt.imshow(img[128,:,:].T,cmap='gray')
    plt.axis('off')
    plt.title("X")
    
    plt.subplot(2, 3, 2)
    plt.imshow(img[:,128,:].T,cmap='gray')
    plt.axis('off')
    plt.title("Y")
    
    plt.subplot(2, 3, 3)
    plt.imshow(img[:,:,128].T,cmap='gray')
    plt.axis('off')
    plt.title("Z")
    
    
    plt.subplot(2, 3, 4)
    plt.imshow(mask[128,:,:].T,interpolation='nearest')
    plt.axis('off')
    plt.title("Mask X")
    
    plt.subplot(2, 3, 5)
    plt.imshow(mask[:,128,:].T,interpolation='nearest')
    plt.axis('off')
    plt.title("Mask Y")
    
    plt.subplot(2, 3, 6)
    plt.imshow(mask[:,:,128].T,interpolation='nearest')
    plt.axis('off')
    plt.title("Mask Z")
plt.show()
```
